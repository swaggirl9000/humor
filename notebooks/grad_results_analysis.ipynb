```python
results = torch.load("/home/ada/humor/data/input_gradient.pt", map_location=torch.device('cpu'))
```

```python
target_attributions = {}
for k, v_list in results.items():
    target_attributions[k] = []
    for v in v_list:
        attributions = v.sequence_attributions[0].target_attributions.sum(-1)
        target_attributions[k].append(attributions)
```
