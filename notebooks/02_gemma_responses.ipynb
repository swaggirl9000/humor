```python
from collections import defaultdict
```

```python
results_dict = defaultdict(list)
for index, row in transcripts.iterrows():
    comedian = row['comedian']
    transcript = row['transcript'] 
    
    prompt = f"'''{transcript}'''\n\n{instruction}. Explain why they are funny?"
    input_ids = tokenizer(prompt, return_tensors="pt").input_ids.to("cuda")
    outputs = model.generate(input_ids=input_ids, max_new_tokens=350)
    generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
    response = generated_text.replace(prompt, "").strip()
    results_dict[comedian].append(response)
```

```python
for comedian, result in results_dict.items():
    print(comedian)
    print(result)
    print()
```
