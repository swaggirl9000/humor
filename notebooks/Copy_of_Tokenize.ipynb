```python
def update(laugh_times):
  updated_laugh_times = [
      {'start': round(laugh['start'], 3), 'end': round(laugh['end'], 3)}
      for laugh in laugh_times
  ]

  return updated_laugh_times
```

```python
import json
with open('dg_output.json', 'r') as file:
    transcript_data = json.load(file)
```

```python
def create_laughter_map(laugh_times, transcript_json):
    laughter_map = {}
    for i, laugh in enumerate(laugh_times):
        laugh_key = f"laugh {i+1}"
        start_laugh = laugh['start']
        end_laugh = laugh['end']
        matching_sentences = set()
        for fragment in transcript_json:
            start_frag = float(fragment['begin'])
            end_frag = float(fragment['end'])
            if start_frag <= end_laugh and end_frag >= start_laugh:
                sentence = " ".join(fragment['lines']).strip()
                matching_sentences.add(sentence)

        laughter_map[laugh_key] = {
            'begin': start_laugh,
            'end': end_laugh,
            'sentences': list(matching_sentences)
        }
    return laughter_map
```

```python
def perform_attribution(model, tokenizer, input_text, generated_text, layer):
    attrib_model = inseq.load_model(
        model,
        "layer_gradient_x_activation",
        tokenizer=MODEL_ID,
        target_layer=model.model.layers[layer].mlp,
    )
    out = attrib_model.attribute(
        input_text,
        generated_text,
        attributed_fn="contrast_prob_diff",
        step_scores=["contrast_prob_diff"],
        show_progress=False,
    )
    # Save aggregated attributions to disk
    out = out.aggregate()
    out.save(f"layer_{layer}_funny_parts.json", overwrite=True)

for layer in range(num_layers):
    perform_attribution(model, tokenizer, input, output, layer)
```
