```python
import seaborn as sns
import torch
import pandas as pd
```

```python
x1 = pd.DataFrame(torch.rand(100) / 3).assign(model="random") 
x2 = pd.DataFrame(torch.rand(100) / 2).assign(model="brute force")
x3 = pd.DataFrame(torch.rand(100) * 2).assign(model="LLM") 

df = pd.concat([x1, x2, x3])

sns.displot(df, x=0, hue="model", kind="kde")
```
